# Bineary Class Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sample data\\heart.csv")
df
```

```python
# x = df.drop(['target'],axis=1)   # this line also work like below line
x = df.iloc[:,:-1]

# y = df[['target']]     # this line also work like below line
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
len(x_train), len(x_test) , len(y_train), len(y_test) 
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score


# f1_score check karte hai in classificaion and it is similer to accuracy_score
```

```python
print("test dataset accuracy_score : ",accuracy_score(y_test, y_pred)), 
```

```python
y_pred1 = model.predict(x_train)
y_pred1
```

```python
print("train dataset accuracy_score : ",accuracy_score(y_train, y_pred1)), 
```

## f1_score

```python
print("test dataset f1_score : ",f1_score(y_test, y_pred)), 
```

```python
print("train dataset f1_score : ",f1_score(y_train, y_pred1)), 
```

## confusion_matrix

```python
confusion_matrix(y_test,y_pred)
```

```python
print("Confusion metrix On test data")
print()
pd.DataFrame(confusion_matrix(y_test,y_pred),columns=list(range(0,2)))
```

```python
print("Confusion metrix On train data")
print()
pd.DataFrame(confusion_matrix(y_train,y_pred1),columns=list(range(0,2)))
```

```python
from sklearn.metrics import recall_score, precision_score



print("For Logistic Regression Model")
print("-"*50)
cf = pd.DataFrame(confusion_matrix(y_test,y_pred),columns=list(range(0,2)))
print(cf)
print("-"*50)

print("test dataset accuracy_score : ",accuracy_score(y_test, y_pred)), 
print("test dataset f1_score : ",f1_score(y_test, y_pred)), 
print("test dataset recall_score : ",recall_score(y_test, y_pred)), 
print("test dataset precision_score : ",precision_score(y_test, y_pred)),
```
